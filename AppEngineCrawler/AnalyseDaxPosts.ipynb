# Analyse DAX Posts

```python
#Load file function
def load_file(filename):
    f = open(filename,'r')
    data = f.read()
    f.close()
    return data
```

### Load and manipulate data

```python
import json
import pandas as pd

#Set display properties
pd.set_option("display.line_width",500)
pd.set_option("display.width",1000)
#Load all WKN posts
messages = json.loads(load_file('db/dax_filtered.txt'))
#Transfer to pandas dataframe
db = pd.DataFrame.from_records(messages)
#Map name short
db['from_name'] = db['from'].map(lambda x: x['name'])
#Create time index
db = db.set_index(pd.to_datetime(db['created_time']))
#Sort for time when post was created
db = db.sort_index(ascending=False)
```

```python
# Helper function for printing out single messages
def print_msg(row):
    print row['created_time']
    print row['message']
    if not pd.isnull(row['comments']):
        for ix, cmt in enumerate(row['comments']['data']):
            print 'Comment {}: '.format(ix+1) + cmt['message']
```

### Set Long/Short, buy and stop

```python
import re
# Set long/short
long_re = re.compile(r'^dax\s*long\s', re.IGNORECASE)
db['longshort'] = db['message'].apply(lambda _: 'long' if long_re.findall(_) else 'short')
# Set trade start
in_re = re.compile(r'\s[0-9]{4,5}\s', re.IGNORECASE)
db['in'] = db['message'].apply(lambda _: int(in_re.search(_).group()[1:-1]))
# Set stop
stop_re = re.compile(r'stop\s*([0-9]{4,5}|([0-9]{1,2}\spk))', re.IGNORECASE)
stop_re_std = re.compile(r'[0-9]{4,5}', re.IGNORECASE)
stop_re_pk = re.compile(r'[0-9]{1,2}\spk', re.IGNORECASE)
def get_stop(msg):
    s1 = stop_re.search(msg).group()
    s2 = stop_re_std.search(s1)
    s3 = stop_re_pk.search(s1)
    if s2:
        return int(s2.group())
    elif s3:
        return int(s3.group()[:-len(' pk')])
    else:
        raise Exception('Stop not found in msg ' + msg)

db['stop'] = db['message'].apply(get_stop)
for idx, stop in enumerate(db['stop']):
    if stop < 100:
        # The stop is denoted as distance from trade start
        db['stop'][idx] = (db['in'][idx] - stop) if (db['longshort'][idx] == 'long') else (db['in'][idx] + stop)
            
```

```python
import numpy as np
einstand_re = re.compile(r'(?:^|\s|$)einstand(?:^|\s|$)', re.IGNORECASE)
tvk_re = re.compile(r'(?:^|\s|$)(tvk|tp|[0-9](st|nd|rd|th))(?:^|\s|$)', re.IGNORECASE)
index_re = re.compile(r'[0-9]{4,5}')
def get_tradeout(row, num_tvk):
    einstand = False
    tvks = []
    for com in row['comments']['data']:
        if einstand_re.findall(com['message']):
            einstand = True
        if tvk_re.findall(com['message']):
            ix = index_re.search(com['message'])
            if ix:
                tvks.append(int(ix.group()))
                einstand = True
                
    tvks.append(row['in'] if einstand else row['stop'])
    return np.mean(tvks[:min([num_tvk, len(tvks)])])
                      
db['out'] = db.apply(lambda row: get_tradeout(row, 2), axis=1)
db['pl'] = db.apply(lambda row: row['out'] - row['in'] if row['longshort'] == 'long' else row['in'] - row['out'], axis=1)
#add commission
db['pl'] += db['pl'].apply(lambda _: -3 if _ <= 0 else -6)

db.sort_index(ascending=True, inplace=True)
jan13 = db.ix['2013':'2014']
jan13['pl_cs'] = jan13['pl'].cumsum()
jan13[['pl', 'pl_cs']].plot()
```
