# Database Pre-Filtering

Dieses Skript lädt die Listen aus db1.txt - db12.txt und extrahiert alle Posts die eine WKN-Nummer enthalten. Diese Posts werden dann separat in einem einzelnen File gespeichert.

```python
#Load file function
def load_file(filename):
    f = open(filename,'r')
    data = f.read()
    f.close()
    return data

def write_file(filename, content):
    f = open(filename,'w')
    f.write(content)
    f.close()
```

```python
import json
import re

#Create Regex compiler for WKN numbers
dax = re.compile(r'^dax\s(short|long)\s*[0-9]{4,5}\s*stop\s*([0-9]{4,5}|([0-9]{1,2}\spk))', re.IGNORECASE)
wkn = re.compile(r'(?:^|\s|$)(?=[^\s]{0,5}[A-Z])(?=[^\s]{0,5}[0-9])[A-Z0-9]{6}(?:^|\s|$)')

wkn_list = [] #Keeps final set of messages
for db in range(1,13):
    #Problem with db6.txt (incorrect JSON format)
    if db == 6:
        continue
    print 'Extracting file number {}'.format(db)
    #Open db[x].txt file and decrypt JSON
    data = load_file('db/db{}.txt'.format(db))
    messages = json.loads(data)
    #Find all messages including a WKN number and store in new list
    for msg in messages:
        if "message" in msg.keys() and "from" in msg.keys() and msg["from"]["name"] == "Orkan Kuyas" and \
                dax.findall(msg["message"]) and not wkn.findall(msg["message"]):
            wkn_list.append(msg)

print 'Writing to DAX file'
write_file('db/dax_filtered.txt', json.dumps(wkn_list))
print 'Done'
```
